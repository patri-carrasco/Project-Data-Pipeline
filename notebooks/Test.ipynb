```python
def pounds_euro(l):
    """
        Función que convierte las libras a euros
        arg:
            l, valor en libras que se quiere calcular
        return
            valor en euros con dos decimales
        
        puntos a mejor: 
          usar api https://cambio.today/api
    """
    return round(l*1.11,2)
```

```python
def price_string_to_price_euro(string):
    try:
        new_string = string[1:]
        price = float(new_string)
    except:
        price = 0
    
    return price
```

```python
euro = []
for  elem in data.price:
    p =price_string_to_price_euro(elem)
    euro.append(pounds_euro(p))

    
data['euro'] = euro
```
